```python
import random
def createChromosome(n):
    ch=""
    for i in range(n):
        ch=ch+str(random.randint(0,n-1))
    return ch

def chromosomeToBoard(ch):
    b=[]
    for i in range(len(ch)):
        row=[]
        for i in range(len(ch)):
            row.append(0)
        b.append(row)

    for i in range(len(ch)):
        r=int(ch[i])
        b[r][i]=1
    return b

def printBoard(b):
    for i in b:
        print(i)
        
def findAttack(ch):
    attack=0
    for i in range(len(ch)):
        for j in range(i+1,len(ch)):
            if ch[i]==ch[j]:
                attack=attack+1
            if abs(i-j)==abs(int(ch[i])-int(ch[j])):
                attack=attack+1
    return attack

def nonAttacking(ch):
    n=len(ch)
    return int((n*(n-1)/2)-findAttack(ch))

def createPopulation(psize,n):
    pop=[]
    for i in range(psize):
        ch=createChromosome(n)
        #print(ch)
        #b=chromosomeToBoard(ch)
        #printBoard(b)
        #print(findAttack(ch))
        #print(nonAttacking(ch))
        pop.append(ch) 
    return pop

def fitness(p):
    fit={}
    for i in range(len(p)):
        ch=p[i]
        fit[ch]=nonAttacking(ch)
    return fit
    
 
psize=4
n=8
pop=createPopulation(psize,n)
#print(pop)
f=fitness(pop)
print(f)
```
