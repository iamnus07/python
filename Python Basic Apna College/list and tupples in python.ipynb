```python
student = ["umme", "habiba", "islam", "rupa"];
print (student)

student[3]= "rupu"
print (student)

# list slicing 
print (student[1:3])  # it will print from index 1 to index 2 but not index 3
print (student[:4])   # it will print from starting to index 3
print (student[2:])   # it will print from index 2 to the end
print (student[-4:-1]) # it will print from index -4 to index -2 but not index -1
```

```python
# list method 

list = [1,2,3,4,5];
list.append(6)  # it will add 6 at the end of the list' 
print (list)
list.insert(2, 2.5) # it will insert 2.5 at index 2
print (list)
list.remove(3)  # it will remove 3 from the list
print (list)
list.pop()  # it will remove the last element from the list
print (list)
list.sort()  # it will sort the list in ascending order
print (list)
list.sort(reverse=True)  # it will sort the list in descending order
print (list)
list.reverse()  # it will reverse the list
print (list)
```

```python
# tupples are immutable means we cannot change the value of a tupple
# tupples do not have any method like lists
# we cannot add or remove elements from a tupple

tup= (1,2,3,1);
tup.index(2)  # it will return the index of 2 in the tupple
print (tup.index(2))

tup.count(1)  # it will return the count of 1 in the tupple
print (tup.count(1))
```
