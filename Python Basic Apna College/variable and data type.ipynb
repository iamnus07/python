```python
# types og operators

# arithmetic operators

a = 32;
b = 8;

sum = a +b;
print ("the sum of a and b is ", sum)

diff = a - b;
print ("the difference of a and b is ", diff)

prod = a * b;
print ("the product of a and b is ", prod)

div = a / b;
print ("the division of a and b is ", div)

floor = a // b;
print ("the floor division of a and b is ", floor)

mod = a % b;
print ("the modulus of a and b is ", mod)

exp = a ** b;
print ("the exponent of a and b is ", exp)
```

```python
# reational operator 

a = 50;
b = 20;

print ("a is greater than b: ", a > b)
print ("a is less than b: ", a < b)
print ("a is equal to b: ", a == b)
print ("a is not equal to b: ", a != b)
print ("a is greater than or equal to b: ", a >= b)
print ("a is less than or equal to b: ", a <= b)
```

```python
# assignment operators
a = 10
b = 5

a += b  # a = a + b
print("After a += b, a =", a)
a -= b  # a = a - b
print("After a -= b, a =", a)
a *= b  # a = a * b
print("After a *= b, a =", a)
a /= b  # a = a / b   
print("After a /= b, a =", a)
a %= b  # a = a % b
print("After a %= b, a =", a)
a **= b  # a = a ** b
print("After a **= b, a =", a)
a //= b  # a = a // b
print("After a //= b, a =", a)
```

```python
# logical operators 

a = 50;

b = 30;

print ("a > 40 and b < 40: ", a > 40 and b < 40)
print ("a > 60 or b < 40: ", a > 60 or b < 40)
print ("not (a > 40): ", not (a > 40)) 
```
