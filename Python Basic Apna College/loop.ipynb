```python
# search for a number X in the list

list = [1,4, 9,16,25,36,49,64,81,100]
X = int(input("Enter a number to search in the list: "))

i = 0;
while i < len(list):
    if list[i] == X:
        print(f"Number {X} found in the list.")
        break
       
    else:
        print(f"Number {X} not found at index {i}.")
    i+=1
```

```python
# break:
i = 1
while i <= 10:
    if i == 6:
        break
    print(i)
    i += 1  
```

```python
# continue:
i = 0
while i < 10:
    i += 1
    if i == 5:
        continue
    print(i)
```
