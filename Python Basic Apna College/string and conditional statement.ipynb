```python
# indexing

str1 = "umme habiba islam rupa"

print(str1[6]);

# with the help of indexing we can access each character of a string but cannot change it because strings are immutable
```

```python
# slicing 

str1 = "umme habiba islam rupa"

print(str1 [0:5]);

print (str1 [5 : len(str1)]);

print (str1[5:])

print (str1 [:5])
```

```python
# negative indexing 

str1 = "apple";
print (str1 [-5:-2])
```

```python
# conditional statement ::

marks = float (input ("please enter your marks : "))

if marks >=90 and marks <=100:
    print("A+")
elif marks >=80 and marks <90:
    print("A")
elif marks >=70 and marks <80:
    print("B")
elif marks >=60 and marks <70:
    print("C")
elif marks >=50 and marks <60:
    print("D")
else:
    print("F")
```

```python
# nesting conditional statement ::
age = int (input ("please enter your age : "))

if age >=18:
    print("you are eligible for voting")
    if age >=21:
        print("you are also eligible for driving")
    else:
        print("you are not eligible for driving")
```
