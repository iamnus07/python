```python
# Problem 6:
# Take an integer number as input from the user
# Reverse the digits of the number
# Print the reversed number


num = int(input("Enter an integer number: "))


reverse_num = 0


while num != 0:
    digit = num % 10
    reverse_num = reverse_num * 10 + digit
    num = num // 10


print("Reversed number:", reverse_num)
```

```python
# Problem 7:
# Take an integer number as input from the user
# Check whether the number is a palindrome
# Print the result


num = int(input("Enter an integer number: "))


original_num = num


reverse_num = 0
while num != 0:
    digit = num % 10
    reverse_num = reverse_num * 10 + digit
    num = num // 10


if original_num == reverse_num:
    print("The number is a Palindrome")
else:
    print("The number is Not a Palindrome")

```

```python
# Problem 8:
# Take a positive integer as input from the user
# Check whether the number is a prime number
# Print the result


num = int(input("Enter a positive integer: "))

if num <= 1:
    print("Not a Prime Number")
else:
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print("Prime Number")
    else:
        print("Not a Prime Number")
```

```python
# Problem 9:
# Take a positive integer N as input from the user
# Print all prime numbers between 1 and N


n = int(input("Enter a positive integer: "))


for num in range(2, n + 1):
    is_prime = True

    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num)

```

```python
# Problem 10:
# Take an integer number as input from the user
# Count the number of digits in the number
# Print the result


num = int(input("Enter an integer number: "))

count = 0
temp = abs(num)  


if temp == 0:
    count = 1
else:
    while temp != 0:
        temp = temp // 10
        count += 1

print("Number of digits:", count)

```

```python
# Problem 12:
# Take a positive integer N as input from the user
# Print the first N terms of the Fibonacci sequence


n = int(input("Enter the number of terms: "))


a, b = 0, 1


if n <= 0:
    print("Please enter a positive integer")
elif n == 1:
    print(a)
else:
    print(a, b, end=" ")
    for _ in range(2, n):
        c = a + b
        print(c, end=" ")
        a, b = b, c

```

```python
# Problem 13:
# Take two positive integers as input from the user
# Find and print their Greatest Common Divisor (GCD)


a = int(input("Enter first number: "))
b = int(input("Enter second number: "))


x, y = a, b
while y != 0:
    x, y = y, x % y

print("GCD of", a, "and", b, "is:", x)
```

```python
# Problem 14:
# Take two positive integers as input from the user
# Find and print their Least Common Multiple (LCM)


a = int(input("Enter first number: "))
b = int(input("Enter second number: "))


def gcd(x, y):
    while y != 0:
        x, y = y, x % y
    return x

lcm = (a * b) // gcd(a, b)


print("LCM of", a, "and", b, "is:", lcm)
```

```python
# Problem 16:
# Take an integer number as input from the user
# Check whether the number is an Armstrong number
# (Armstrong number: sum of its digits each raised to the power of number of digits equals the number itself)
# Print the result


num = int(input("Enter a number: "))

original_num = num

num_digits = len(str(abs(num))) 


sum_digits = 0
temp = abs(num)
while temp != 0:
    digit = temp % 10
    sum_digits += digit ** num_digits
    temp = temp // 10

if sum_digits == abs(original_num):
    print(f"{original_num} is an Armstrong number")
else:
    print(f"{original_num} is not an Armstrong number")
```

```python
# Problem 19:
# Take a positive integer N as input from the user
# Print a right-angled triangle pattern of stars (*) with N rows
# Example for N = 5:
# *
# **
# ***
# ****
# *****


n = int(input("Enter the number of rows: "))

for i in range(1, n + 1):
    print("*" * i)
```

```python
# Problem 20:
# Take a positive integer N as input from the user
# Print an inverted right-angled triangle pattern of stars (*) with N rows
# Example for N = 5:
# *****
# ****
# ***
# **
# *


n = int(input("Enter the number of rows: "))

for i in range(n, 0, -1):
    print("*" * i)

```

```python
# Problem 21:
# Take a positive integer N as input from the user
# Print a centered pyramid pattern of stars (*) with N rows
# Example for N = 5:
#     *
#    ***
#   *****
#  *******
# *********


n = int(input("Enter the number of rows: "))


for i in range(1, n + 1):
    
    print(" " * (n - i), end="")
    
    print("*" * (2 * i - 1))
```
