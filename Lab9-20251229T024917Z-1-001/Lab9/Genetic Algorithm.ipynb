```python
import random
def createChromosome(n):
    ch=""
    for i in range(n):
        ch=ch+str(random.randint(0,n-1))
    return ch

def chromosomeToBoard(ch):
    b=[]
    for i in range(len(ch)):
        row=[]
        for i in range(len(ch)):
            row.append(0)
        b.append(row)

    for i in range(len(ch)):
        r=int(ch[i])
        b[r][i]=1
    return b

def printBoard(b):
    for i in b:
        print(i)
        
def findAttack(ch):
    attack=0
    for i in range(len(ch)):
        for j in range(i+1,len(ch)):
            if ch[i]==ch[j]:
                attack=attack+1
            if abs(i-j)==abs(int(ch[i])-int(ch[j])):
                attack=attack+1
    return attack

def nonAttacking(ch):
    n=len(ch)
    return int((n*(n-1)/2)-findAttack(ch))

def createPopulation(psize,n):
    pop=[]
    for i in range(psize):
        ch=createChromosome(n)
        #print(ch)
        #b=chromosomeToBoard(ch)
        #printBoard(b)
        #print(findAttack(ch))
        #print(nonAttacking(ch))
        pop.append(ch) 
    return pop

def fitness(p):
    fit={}
    for i in range(len(p)):
        ch=p[i]
        fit[ch]=nonAttacking(ch)
    return fit

def selection(pf):
    #sort the dictionary
    spf=sorted(pf.items(), key=lambda x:x[1])
    #print(spf)
    parent1=spf[-1][0]
    parent2=spf[-2][0]
    return parent1,parent2

def crossover(p1,p2):
    point=random.randint(0,len(p1)-1)
    #print(point)
    ch1=p1[:point]+p2[point:]
    ch2=p2[:point]+p1[point:]
    return ch1,ch2

def mutation(p1,p2):
    point=random.randint(0,len(p1)-1)
    value=random.randint(0,len(p1)-1)
    op1=p1[:point]+str(value)+p1[point+1:]
    op2=p2[:point]+str(value)+p2[point+1:]
    return op1,op2
 
psize=4
n=10
pop=createPopulation(psize,n)
#print(pop)
for i in range(100):
    f=fitness(pop)
    #print(f)
    p1,p2=selection(f)
    #print(p1,p2)
    child1,child2=crossover(p1,p2)
    #print(child1,child2)
    offspring1,offspring2=mutation(child1,child2)
    opList=[offspring1,offspring2]
    cf=fitness(opList)
    print(cf)
    pop.append(offspring1)
    pop.append(offspring2)
    new_gen=fitness(pop)

print(new_gen)
```
