```python
def stepFunction(start=0,stepSize=0.1,n=10):
    s=[]
    for i in range(n):
        s.append(start)
        start=start+stepSize
    return s

list=stepFunction(-1,0.1,10)
print(list)
```

```python
import matplotlib.pyplot as plt
def f(x):
    return x**2
def linspace(a=0,b=1,n=10):
    s=[]
    del_h=(b-a)/n
    #print(del_h)
    s.append(a)
    for i in range(n):
        a+=del_h
        s.append(a)
    return s
x=linspace(-10,10,200)
y=[]
for i in range(len(x)):
    y.append(f(x[i]))
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
def createMatrix(n):
    m=[]
    for i in range(n):
        c=[]
        for j in range(n):
            c.append(0)
        m.append(c)
    return m

def printMatrix(m):
    for r in m:
        print(r)

g=createMatrix(6)
g[0][1]=1
g[0][2]=1
g[1][2]=1
g[1][3]=1
g[2][3]=1
g[3][4]=1
g[4][0]=1
g[4][1]=1
g[4][5]=1
printMatrix(g)
```
