```python
# dictionary methods
student = {
    "name":"umme habiba islam rupa",
    "age":22,
    "marks":{
        "cse":85,
        "math":90,
        "phy":75,
        "chem":80,
        "english":88
    }
}

print (student.keys())
print (student.values())
print (student.items())

print (list (student.keys()))

print (list (student.values())) 
print (list (student.items()))

print (student.get ("name"))

```

```python
# set methods
nums= {1, 2, 3, 4, 5}

nums.add (6)
print (nums)
nums.add("shuvo")
print (nums)
nums.add ((7,8,9))
print (nums)
nums.remove (3)
print (nums)
nums.discard (10)  # does not raise an error if the element is not found
print (nums)
nums.pop()  # removes and returns an arbitrary element from the set
print (nums)

nums.clear()  # removes all elements from the set
print (nums)


# union method 
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}
set3 = set1.union (set2)
print (set3)

# intersection method
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}
set3 = set1.intersection (set2)
print (set3)
```
