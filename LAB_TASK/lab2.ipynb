```python
# Problem:1
# 1. Create a list of numbers
# 2. Remove duplicates
# 3. Find maximum, minimum, and average manually (without using built-in max(), min(), sum())


numbers = [4, 2, 8, 2, 10, 4, 5]

unique_numbers = list(set(numbers))
print("List after removing duplicates:", unique_numbers)


max_num = unique_numbers[0]
for num in unique_numbers:
    if num > max_num:
        max_num = num
print("Maximum number:", max_num)

min_num = unique_numbers[0]
for num in unique_numbers:
    if num < min_num:
        min_num = num
print("Minimum number:", min_num)

total = 0
for num in unique_numbers:
    total += num
average = total / len(unique_numbers)
print("Average:", average)
```

```python
# Problem:2
# 1. Given a list of integers
# 2. Create a new list containing only numbers divisible by 3

numbers = [1, 3, 4, 6, 7, 9, 12, 14, 15]

div_by_3 = []

for num in numbers:
    if num % 3 == 0:
        div_by_3.append(num)

print("Numbers divisible by 3:", div_by_3)

```

```python
# Problem:4
# 1. Given two lists: [1,2,3] and [4,5,6]
# 2. Merge them into a single list
# 3. Remove the middle element
# 4. Print the updated list

list1 = [1, 2, 3]
list2 = [4, 5, 6]

merged_list = list1 + list2
print("Merged list:", merged_list)

middle_index = len(merged_list) // 2 

removed_element = merged_list.pop(middle_index)

print("Removed middle element:", removed_element)
print("Updated list:", merged_list)
```

```python
# Problem:6
# 1. Given a dictionary:
#    student = {"name": "Rafi", "marks": [70, 80, 90]}
# 2. Add a new subject mark (e.g., 85)
# 3. Calculate the new average of marks
# 4. Print updated dictionary and average


student = {"name": "Rafi", "marks": [70, 80, 90]}
print("Original student dictionary:", student)


new_mark = 85
student["marks"].append(new_mark)
print("Updated student dictionary:", student)

total = 0
for mark in student["marks"]:
    total += mark
average = total / len(student["marks"])

print("New average of marks:", average)

```

```python
# Problem:7
# 1. Create a dictionary of fruits: {"apple": 1, "banana": 2, "mango": 3}
# 2. Create another dictionary where keys and values are swapped
# 3. Print the new dictionary

fruits = {"apple": 1, "banana": 2, "mango": 3}
print("Original dictionary:", fruits)

swapped_fruits = {value: key for key, value in fruits.items()}

print("Swapped dictionary:", swapped_fruits)
```

```python
# Problem:8
# 1. Take an input string from the user
# 2. Count the occurrence of each character
# 3. Store the counts in a dictionary
# 4. Print the dictionary
# Example: "hello" -> {'h':1, 'e':1, 'l':2, 'o':1}



text = input("Enter a string: ")

char_count = {}

for char in text:
    if char in char_count:
        char_count[char] += 1  
    else:
        char_count[char] = 1   


print("Character count:", char_count)
```

```python
# Problem:11
# 1. Write a function that calculates simple interest
# 2. Formula: SI = (p * r * t) / 100
# 3. r (rate) = 5% by default if not provided
# 4. p = principal, t = time (in years)
# 5. Return the simple interest



def simple_interest(p, t, r=5):
    """
    Calculates Simple Interest
    p: Principal amount
    t: Time in years
    r: Rate of interest (default 5%)
    """
    si = (p * r * t) / 100
    return si


principal = 1000
time = 3


si_default = simple_interest(principal, time)
print(f"Simple Interest (default rate 5%): {si_default}")


custom_rate = 7
si_custom = simple_interest(principal, time, custom_rate)
print(f"Simple Interest (custom rate {custom_rate}%): {si_custom}")
```

```python
# Problem:12
# 1. Write a function that takes two numbers as input
# 2. Returns both their sum and difference
# 3. Print the results


def sum_and_difference(a, b):
    """
    Returns sum and difference of two numbers
    """
    sum_result = a + b
    diff_result = a - b
    return sum_result, diff_result  


num1 = 10
num2 = 4

sum_val, diff_val = sum_and_difference(num1, num2)

print(f"Sum of {num1} and {num2} is: {sum_val}")
print(f"Difference of {num1} and {num2} is: {diff_val}")
```

```python
# Problem:13
# 1. Write a function that takes a list of numbers
# 2. Returns a new list containing the square of each number


def square_list(numbers):
    """
    Takes a list of numbers and returns a new list with their squares
    """
    squares = []
    for num in numbers:
        squares.append(num ** 2)
    return squares


nums = [1, 2, 3, 4, 5]
squared_nums = square_list(nums)

print("Original list:", nums)
print("Squared list:", squared_nums)

```

```python
# Problem:14
# 1. Take 5 names from the user
# 2. Write them to a text file
# 3. Read the file and print all names


names = []
for i in range(5):
    name = input(f"Enter name {i+1}: ")
    names.append(name)


with open("names.txt", "w") as file:
    for name in names:
        file.write(name + "\n")  

print("Names written to 'names.txt' successfully.")


print("\nReading names from file:")
with open("names.txt", "r") as file:
    for line in file:
        print(line.strip()) 
```

```python
# Problem:15
# 1. Open a text file
# 2. Count the number of lines in the file
# 3. Print the total line count


filename = "names.txt"  

line_count = 0

with open(filename, "r") as file:
    for line in file:
        line_count += 1 


print(f"Total number of lines in '{filename}': {line_count}")

```

```python
# Problem:16
# 1. Generate a random list of 10 integers between 1 and 50
# 2. Print the list
# 3. Find and print the highest and lowest numbers


import random


random_numbers = [random.randint(1, 50) for i in range(10)]
print("Random list:", random_numbers)


highest = random_numbers[0]
lowest = random_numbers[0]

for num in random_numbers:
    if num > highest:
        highest = num
    if num < lowest:
        lowest = num

print("Highest number:", highest)
print("Lowest number:", lowest)

```

```python
# Problem:17
# 1. Ask the user for a number
# 2. Print its square root
# 3. Print its factorial
# 4. Print its cube (power of 3)


import math


num = int(input("Enter a number: "))


sqrt_num = math.sqrt(num)


factorial_num = math.factorial(num)


cube_num = num ** 3


print(f"Square root of {num}: {sqrt_num}")
print(f"Factorial of {num}: {factorial_num}")
print(f"{num} raised to power 3: {cube_num}")
```
