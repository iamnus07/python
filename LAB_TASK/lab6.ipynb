```python
# Define the graph using an adjacency list
graph = {
    0: [1, 2],
    1: [3],
    2: [4],
    3: [],
    4: [5],
    5: []
}

# Initialize a set to track visited nodes
visited = set()

# Define DFS function
def dfs(node):
    print(node, end=' ')       # Visit the node
    visited.add(node)          # Mark as visited
    for neighbor in graph[node]:
        if neighbor not in visited:
            dfs(neighbor)

# Call DFS starting from node 0
print("DFS Traversal starting from node 0:")
dfs(0)
```

```python
# Graph (adjacency list)
G = {
    0: [1, 2],
    1: [3],
    2: [4],
    3: [],
    4: [5],
    5: []
}

# Set to track visited nodes
visited = set()

# Recursive DFS function
def dfs_recursive(G, start):
    print(start, end=' ')      # Visit the node
    visited.add(start)         # Mark as visited
    for neighbor in G[start]:
        if neighbor not in visited:
            dfs_recursive(G, neighbor)

# Call DFS starting from node 0
print("DFS Traversal (Recursive) starting from node 0:")
dfs_recursive(G, 0)
```

```python
# Graph (adjacency list)
G = {
    0: [1, 2],
    1: [3],
    2: [4],
    3: [],
    4: [5],
    5: []
}

# Recursive DFS function with path tracking
def dfs_path(G, node, visited=None, path=None):
    if visited is None:
        visited = set()
    if path is None:
        path = []

    visited.add(node)
    path.append(node)  # Add current node to path

    # Print path to the current node
    print("Path to", node, ":", " â†’ ".join(map(str, path)))

    # Explore neighbors
    for neighbor in G[node]:
        if neighbor not in visited:
            dfs_path(G, neighbor, visited, path.copy())  # Use path.copy() to avoid modifying other paths

# Call DFS starting from node 0
print("DFS Traversal with Paths:")
dfs_path(G, 0)
```

```python
from queue import Queue

def bfs_adj_list(G, start):
    visited = set()
    q = Queue()
    
    visited.add(start)
    q.put(start)
    
    print("BFS Traversal (Adjacency List):", end=' ')
    
    while not q.empty():
        node = q.get()
        print(node, end=' ')
        
        for neighbor in G[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                q.put(neighbor)

# Example graph (adjacency list)
G_list = {
    0: [1, 2],
    1: [3],
    2: [4],
    3: [],
    4: [5],
    5: []
}

bfs_adj_list(G_list, 0)
```
