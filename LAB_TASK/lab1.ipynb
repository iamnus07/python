```python
# Problem 16:
# Take a score (0-100) as input
# Assign a grade based on the score:
# A: 90-100
# B: 80-89
# C: 70-79
# F: <70
# Print the grade


score = float(input("Enter the score (0-100): "))


if 90 <= score <= 100:
    grade = "A"
elif 80 <= score < 90:
    grade = "B"
elif 70 <= score < 80:
    grade = "C"
else:
    grade = "F"

print("Grade:", grade)

```

```python
# Problem 25:
# Write a function to count even numbers in a list


def count_even(lst):
    count = 0
    for num in lst:
        if num % 2 == 0:
            count += 1
    return count


numbers = [1, 2, 3, 4, 5, 6]
print("Number of even numbers:", count_even(numbers))
```

```python
# Problem 26:
# Write a function to reverse a list without using the reverse() method

def reverse_list(lst):
    return lst[::-1]  # slicing method


numbers = [1, 2, 3, 4, 5]
print("Reversed list:", reverse_list(numbers))
```

```python
# Problem 27:
# Write a function to calculate factorial of a number using recursion
# Function definition


def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)


num = 5
print("Factorial of", num, "is:", factorial(num))
```

```python
# Diamond Pattern:
# Print a diamond pattern using stars (*) with 10 rows as given:
# Example:
#         *
#        ***
#       *****
#      *******
#     **********
#     *********
#      *******
#       *****
#        ***
#         *



n = 5  


for i in range(1, n + 1):
    print(" " * (n - i) + "*" * (2 * i - 1))


for i in range(n, 0, -1):
    print(" " * (n - i) + "*" * (2 * i - 1))
```

```python
# Hollow Square Pattern:
# Print a hollow square using stars (*) with 5 rows as given:
# *****
# *   *
# *   *
# *   *
# *****

# Size of the square

n = 5

for i in range(1, n + 1):
    for j in range(1, n + 1):
        
        if i == 1 or i == n or j == 1 or j == n:
            print("*", end="")
        else:
            print(" ", end="")
    print()  
```
