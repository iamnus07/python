```python
def createMatrix(n):
    m=[]
    for i in range(n):
        c=[]
        for j in range(n):
            c.append(0)
        m.append(c)
    return m

def printMatrix(m):
    for r in m:
        print(r)

f=open("g1.txt","r")
vertex_edges=f.readline()
#print(type(vertex_edges))
vertex_edges_list=vertex_edges.split()
#print(vertex_edges_list)
number_of_vertex=int(vertex_edges_list[0])
number_of_edge=int(vertex_edges_list[1])
#print(number_of_vertex,number_of_edge)

G=createMatrix(number_of_vertex)

for i in range(number_of_edge):
    edge=f.readline()
    edge_list=edge.split()
    u=int(edge_list[0])
    v=int(edge_list[1])
    G[u][v]=1
    G[v][u]=1

printMatrix(G)


def findDegree(G,node):
    count=0
    for i in range(len(G)):
        if G[node][i]==1:
            count=count+1
    return count

def findAllDegrees(G):
    d={}
    for i in range(len(G)):
        d[i]=findDegree(G,i)
    return d


print(findAllDegrees(G))
```
