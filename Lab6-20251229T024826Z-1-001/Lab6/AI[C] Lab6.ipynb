```python
def createMatrix(n):
    m=[]
    for i in range(n):
        c=[]
        for j in range(n):
            c.append(0)
        m.append(c)
    return m

def printMatrix(m):
    for r in m:
        print(r)

f=open("g.txt","r")
vertex_edges=f.readline()

vertex_edges_list=vertex_edges.split()

number_of_vertex=int(vertex_edges_list[0])
number_of_edge=int(vertex_edges_list[1])

G=createMatrix(number_of_vertex)

for i in range(number_of_edge):
    edge=f.readline()
    edge_list=edge.split()
    u=int(edge_list[0])
    v=int(edge_list[1])
    G[u][v]=1

printMatrix(G)

import queue
def dfs(G,s):
    stack=queue.LifoQueue()
    stack.put(s)
    explore_set=set()
    while not stack.empty():
        n=stack.get()
        if n not in explore_set:
            print(n,end=" ")
        explore_set.add(n)
        for i in range(len(G[n])):
            if G[n][i]==1 and i not in explore_set:
                stack.put(i)
dfs(G,0)
```

```python
def createList(n):
    g={}
    for i in range(n):
        g[i]=[]
    return g

def printList(g):
    for k,v in g.items():
        print(k,v)

f=open("g.txt","r")
vertex_edges=f.readline()
vertex_edges_list=vertex_edges.split()
number_of_vertex=int(vertex_edges_list[0])
number_of_edge=int(vertex_edges_list[1])

#print(number_of_vertex,number_of_edge)

G=createList(number_of_vertex)

for i in range(number_of_edge):
    edge=f.readline()
    edge_list=edge.split()
    u=int(edge_list[0])
    v=int(edge_list[1])
    G[u].append(v)
    
printList(G)

import queue
def dfs(G,s):
    stack=queue.LifoQueue()
    stack.put(s)
    explore_set=set()
    while not stack.empty():
        n=stack.get()
        if n not in explore_set:
            print(n,end=" ")
        explore_set.add(n)
        for i in G[n]:
            if i not in explore_set:
                stack.put(i)
dfs(G,0)
```
