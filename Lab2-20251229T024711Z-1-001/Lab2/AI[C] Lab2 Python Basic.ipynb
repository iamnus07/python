```python
def makeDouble(li):
    dList=[]
    for i in range(len(li)):
        dList.append(li[i]*2)
    return dList

list=[10,20,30,40]
dList=makeDouble(list)
print(list)
print(dList)
```

```python
def slope(p):
    x1=p[0][0]
    y1=p[0][1]
    x2=p[1][0]
    y2=p[1][1]
    return (y1-y2)/(x1-x2)

points=[(3,1),(5,6)]
slope(points)
```
