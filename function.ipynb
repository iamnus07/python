```python
# length of a function 
list1 = [1, 2, 3, 4, 5];
def my_len():
    count =0
    for i in list1:
        count += 1
    return count
print (my_len())
```

```python
num = int (input("please enter a number"))

def even_odd_check():
    if(num % 2 ==0):
        print("the number is even")
    else:
        print("the number is odd")

    return even_odd_check

print (even_odd_check())
```

```python
# recursion :
def show (n):
    if(n==0):
        return
    print(n)
    print(n-1)
    print ("end")

show(3)
    
```

```python
def fact (n):
    if (n==0 or n==1):
        return 1;
    else:
        return n* fact(n-1)


fact(3)

```
